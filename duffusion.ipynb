```python
import torch
from diffusers import DiffusionPipeline

# Load model (damo-vilab/text-to-video-ms-1.7b requires <8GB VRAM)
pipe = DiffusionPipeline.from_pretrained(
    "damo-vilab/text-to-video-ms-1.7b",
    torch_dtype=torch.float16,
    variant="fp16"
).to("cuda")

# Enable memory optimizations
pipe.enable_model_cpu_offload()
pipe.enable_vae_slicing()

# Generate video
prompt = "A cat playing piano in a fancy room, 4K HD"
video_frames = pipe(
    prompt,
    num_frames=16,  # 16 frames ≈ 0.67s at 24fps
    num_inference_steps=25,
    height=256,
    width=256
).frames

# Save as MP4
import imageio
imageio.mimsave("output.mp4", video_frames, fps=8, codec="libx264")

```
