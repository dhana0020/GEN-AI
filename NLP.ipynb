SPAM DETECTION

```python


from transformers import pipeline
import gradio as gr

# Load the bert-base-uncased model for text classification
classifier = pipeline("text-classification", model="bert-base-uncased")

# Define label mapping for meaningful output
label_mapping = {"LABEL_0": "spam", "LABEL_1": "ham"}  # Default, adjust as needed

# Define the prediction function
def spam_detection(text):
    result = classifier(text)
    # Map labels based on the model's output
    label = result[0]["label"]
    score = round(result[0]["score"], 2)

    # Attempt to map labels
    if label in label_mapping:
        label = label_mapping[label]
    else:
        # If the label is not in the mapping, print a warning
        print(f"Warning: Unrecognized label {label}. Please update label_mapping.")

    return label, score

# Create and launch the Gradio interface
demo = gr.Interface(
    fn=spam_detection,
    inputs="text",
    outputs=["text", "number"],
    title="Spam Detection App",
    description="Enter text to check if it's spam or not."
)

demo.launch(share=True)
```

```python
from transformers import pipeline
import gradio as gr

# Load a topic classification model (fine-tuned or zero-shot)
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Define the categorization function
def categorize_news(article):
    categories = ["politics", "gaming", "sports", "technology", "entertainment"]
    result = classifier(article, candidate_labels=categories)
    return result['labels'][0], round(result['scores'][0], 2)

# Create and launch the Gradio interface
demo = gr.Interface(
    fn=categorize_news,
    inputs="text",
    outputs=["text", "number"],
    title="Topic Categorization App",
    description="Enter a news article to categorize it into topics."
)

demo.launch(share=True)
```

```python
import gradio as gr
from transformers import pipeline

# Load the text generation model
story_generator = pipeline("text-generation", model="gpt2")

# Define the story generation function
def generate_story(prompt):
    stories = story_generator(prompt, max_length=200, num_return_sequences=1)
    return stories[0]['generated_text']

# Create and launch the Gradio interface
demo = gr.Interface(
    fn=generate_story,
    inputs="text",
    outputs="text",
    title="Story Generation App",
    description="Enter a prompt to generate a story."
)

demo.launch(share=True)
```

```python
import gradio as gr
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Define the story generation function
def generate_story(topic):
    # Create a prompt for the model
    prompt = f"Write an engaging story about {topic}.\n\n"

    # Tokenize input and generate story
    input_ids = tokenizer.encode(prompt, return_tensors='pt')

    # Generate text
    with torch.no_grad():
        output = model.generate(input_ids, max_length=300, num_return_sequences=1, temperature=0.7)

    # Decode generated text
    story = tokenizer.decode(output[0], skip_special_tokens=True)

    return story.strip()

# Create and launch the Gradio interface
demo = gr.Interface(
    fn=generate_story,
    inputs="text",
    outputs="text",
    title="Story Generation App",
    description="Enter a topic to generate a relevant story using GPT-2."
)

demo.launch(share=True)

```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt3"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Define the story generation function
def generate_story(topic):
    # Create a prompt for the model
    prompt = f"Write an engaging story about {topic}.\n\n"

    # Tokenize input and generate story
    input_ids = tokenizer.encode(prompt, return_tensors='pt')

    # Generate text
    with torch.no_grad():
        output = model.generate(
            input_ids,
            max_length=300,  # Limit story length
            num_return_sequences=1,
            temperature=0.7,  # Controls creativity (higher = more creative)
        )

    # Decode generated text
    story = tokenizer.decode(output[0], skip_special_tokens=True)

    return story.strip()

# Main program loop
if __name__ == "__main__":
    print("Welcome to the Story Generation App!")
    print("Enter a topic to generate a story or type 'exit' to quit.")

    while True:
        topic = input("\nEnter a topic: ")
        if topic.lower() == "exit":
            print("Goodbye!")
            break

        # Generate and display the story
        print("\nGenerating story...")
        story = generate_story(topic)
        print("\nGenerated Story:")
        print(story)
```
