```python
# Install required libraries
!pip install requests PyPDF2 python-docx

import os
import re
import requests
import PyPDF2
import docx
from google.colab import files

# üìå Set your Router AI API Key
ROUTER_AI_API_KEY = "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"  # Replace with your key
ROUTER_AI_ENDPOINT = "https://api.router.ai/v1/chat/completions"  # Check your correct endpoint

# üìå Upload Resume (PDF or DOCX)
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Function to extract text from PDF
def extract_text_from_pdf(file_path):
    try:
        with open(file_path, "rb") as f:
            reader = PyPDF2.PdfReader(f)
            text = "\n".join(page.extract_text() for page in reader.pages if page.extract_text())
        return text
    except Exception as e:
        return f"Error reading PDF: {str(e)}"

# Function to extract text from DOCX
def extract_text_from_docx(file_path):
    try:
        doc = docx.Document(file_path)
        text = "\n".join([para.text for para in doc.paragraphs])
        return text
    except Exception as e:
        return f"Error reading DOCX: {str(e)}"

# Detect file type and extract text
if file_name.lower().endswith(".pdf"):
    resume_text = extract_text_from_pdf(file_name)
elif file_name.lower().endswith(".docx"):
    resume_text = extract_text_from_docx(file_name)
else:
    resume_text = "Unsupported file format"

# üìù Define the prompt
prompt_text = f"""
You are a Resume Analyzer AI. Extract the following details:
- Name
- Email
- Phone Number
- Skills
- Experience Summary (2-3 lines)

Resume Text: {resume_text}

Return the extracted details in a structured format.
"""

# üî• Make a request to Router AI API
headers = {
    "Authorization": f"Bearer {ROUTER_AI_API_KEY}",
    "Content-Type": "application/json",
}

data = {
    "model": "gpt-4",  # Replace with the model available in Router AI
    "messages": [{"role": "user", "content": prompt_text}],
    "temperature": 0.2,
}

response = requests.post(ROUTER_AI_ENDPOINT, headers=headers, json=data)

# üìå Print Results
if response.status_code == 200:
    result = response.json()
    print("\nüîπ **Resume Analysis** üîπ")
    print(result["choices"][0]["message"]["content"])
else:
    print("‚ùå Error:", response.text)
```
